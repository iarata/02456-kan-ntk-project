```python
import numpy as np
import pandas as pd
import torch
from torch import nn
import lightning as L
import torch.nn.functional as F
from torchvision.datasets import MNIST
from torch.utils.data import Dataset, DataLoader, Subset
from torchvision import transforms
from lightning.pytorch.loggers import WandbLogger
from introduction_code import GaussianFit, MSELoss_batch
from typing import Union
from sklearn.decomposition import PCA
import sys
sys.path.append('./Convolutional-KANs/kan_convolutional')
from kan import KAN

import matplotlib.pyplot as plt
from datetime import datetime
import json
import os

# Setup Device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device: {device}")

# Setup Randomness
L.seed_everything(628, workers=True)

# CUDA Efficiency
torch.set_float32_matmul_precision('high')

###################################
# Filter MNIST to only include digits 3, 7, 9
###################################
transform = transforms.Compose([
    transforms.ToTensor()
])

allowed_digits_map = {3: 0, 7: 1, 9: 2}  # Original -> New labels

# Load full datasets
train_dataset_full_orig = MNIST("./temp/", train=True, download=True, transform=transform)
test_dataset_orig = MNIST("./temp/", train=False, download=True, transform=transform)

# Filter train dataset
train_indices = [i for i, t in enumerate(train_dataset_full_orig.targets) if t.item() in allowed_digits_map]
train_dataset_full = Subset(train_dataset_full_orig, train_indices)

# Filter test dataset
test_indices = [i for i, t in enumerate(test_dataset_orig.targets) if t.item() in allowed_digits_map]
test_dataset = Subset(test_dataset_orig, test_indices)

# Remap labels in these subsets: We'll create a wrapper dataset for remapping
class RemappedDataset(Dataset):
    def __init__(self, subset, label_map):
        self.subset = subset
        self.label_map = label_map
    def __len__(self):
        return len(self.subset)
    def __getitem__(self, idx):
        x, y = self.subset[idx]
        y = self.label_map[y]
        return x, y

train_dataset_full = RemappedDataset(train_dataset_full, allowed_digits_map)
test_dataset = RemappedDataset(test_dataset, allowed_digits_map)

# Split train_dataset_full into train and validation sets
train_size = int(0.8 * len(train_dataset_full))
val_size = len(train_dataset_full) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(train_dataset_full, [train_size, val_size])

###################################
# Original LCDataset (not used after PCA, but kept for reference)
class LCDataset(Dataset):
    def __init__(self, dataset, num_classes, limit=-1):
        self.limit = limit
        self.num_classes = num_classes
        if self.limit != -1:
            sub = list(np.random.permutation(np.arange(len(dataset)))[0:self.limit]) 
            self.dataset = Subset(dataset, sub)
        else:
            self.dataset = dataset
    
    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        x, y = self.dataset[idx]
        y_one_hot = torch.zeros(self.num_classes)
        y_one_hot[y] = 1
        return x, y_one_hot
###################################

###################################
# PCA Steps
# We'll extract the data from train, val, and test datasets, fit PCA on training, and transform all sets.
###################################
def dataset_to_array(dataset):
    data_list = []
    targets_list = []
    for i in range(len(dataset)):
        x, y = dataset[i]
        # x is [1,28,28] -> Flatten to [784]
        data_list.append(x.numpy().flatten())
        targets_list.append(y)
    data = np.stack(data_list)
    targets = np.array(targets_list)
    return data, targets

# Extract data from train, val, test
train_data, train_targets = dataset_to_array(train_dataset)
val_data, val_targets = dataset_to_array(val_dataset)
test_data, test_targets = dataset_to_array(test_dataset)

# Fit PCA on train data only
n_components = 4  # choose desired number of PCA components
pca = PCA(n_components=n_components)
train_data_pca = pca.fit_transform(train_data)
val_data_pca = pca.transform(val_data)
test_data_pca = pca.transform(test_data)

# Now we create a dataset class for PCA data
class PCADataset(Dataset):
    def __init__(self, data, targets, num_classes=3):
        self.data = data.astype(np.float32)
        self.targets = targets
        self.num_classes = num_classes

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        x = torch.tensor(self.data[idx], dtype=torch.float32)
        y_one_hot = torch.zeros(self.num_classes)
        y_one_hot[self.targets[idx]] = 1
        return x, y_one_hot

pca_train_dataset = PCADataset(train_data_pca, train_targets, num_classes=3)
pca_val_dataset = PCADataset(val_data_pca, val_targets, num_classes=3)
pca_test_dataset = PCADataset(test_data_pca, test_targets, num_classes=3)

def get_pca_dataloader(split="train", batch_size=64, limit=-1):
    if limit != -1:
        if split == "train":
            indices = np.random.permutation(len(pca_train_dataset))[:limit]
            subset = Subset(pca_train_dataset, indices)
            return DataLoader(subset, batch_size=batch_size, shuffle=True)
        elif split == "val":
            indices = np.random.permutation(len(pca_val_dataset))[:limit]
            subset = Subset(pca_val_dataset, indices)
            return DataLoader(subset, batch_size=batch_size, shuffle=False)
        elif split == "test":
            indices = np.random.permutation(len(pca_test_dataset))[:limit]
            subset = Subset(pca_test_dataset, indices)
            return DataLoader(subset, batch_size=batch_size, shuffle=False)
        elif split == "ntk":
            indices = np.random.permutation(len(pca_train_dataset))[:limit]
            subset = Subset(pca_train_dataset, indices)
            return DataLoader(subset, batch_size=batch_size, shuffle=True)

    if split == "train":
        return DataLoader(pca_train_dataset, batch_size=batch_size, shuffle=True)
    elif split == "val":
        return DataLoader(pca_val_dataset, batch_size=batch_size)
    elif split == "test":
        return DataLoader(pca_test_dataset, batch_size=batch_size)
    elif split == "ntk":
        indices = np.random.permutation(len(pca_train_dataset))[:500]
        subset = Subset(pca_train_dataset, indices)
        return DataLoader(subset, batch_size=batch_size)

###################################
# Model Definition
###################################
class ClassicKAN(L.LightningModule):
    def __init__(self, num_hidden_layers=1, hidden_dim=64, inp_size=4, out_size=3, grid_size=2, spline_order=2):
        super().__init__()
        self.inp_size = inp_size
        layers_hidden = [inp_size] + [hidden_dim] * num_hidden_layers + [out_size]
        self.net = KAN(width=layers_hidden, grid=grid_size, k=spline_order)
    
    def forward(self, x):
        return self.net(x)
    
    def training_step(self, batch, batch_idx):
        x, y = batch
        y_pred = self(x)
        loss = F.mse_loss(y_pred, y)  # MSE Loss
        v1 = torch.argmax(y_pred, dim=1)
        v2 = torch.argmax(y, dim=1)
        accuracy = torch.sum(torch.eq(v1, v2)) / len(y)
        self.log("train_loss", loss, prog_bar=True)
        self.log("train_accuracy", accuracy, prog_bar=True)
        return loss

    def validation_step(self, batch, batch_idx):
        x, y = batch
        y_pred = self(x)
        loss = F.mse_loss(y_pred, y)
        v1 = torch.argmax(y_pred, dim=1)
        v2 = torch.argmax(y, dim=1)
        accuracy = torch.sum(torch.eq(v1, v2)) / len(y)
        self.log("val_loss", loss, prog_bar=True)
        self.log("val_accuracy", accuracy, prog_bar=True)

    def test_step(self, batch, batch_idx):
        x, y = batch
        y_pred = self(x)
        loss = F.mse_loss(y_pred, y)
        v1 = torch.argmax(y_pred, dim=1)
        v2 = torch.argmax(y, dim=1)
        accuracy = torch.sum(torch.eq(v1, v2)) / len(y)
        self.log("test_loss", loss)
        self.log("test_accuracy", accuracy)

    def predict_step(self, batch, batch_idx, dataloader_idx=0):
        return self(batch)

    def configure_optimizers(self):
        optimizer = torch.optim.Adam(self.parameters(), lr=1e-4)
        return optimizer

def check_ntk_acc(model, dataloader):
    res = 0.0
    sumlength = 0
    model.eval()
    model.to(device)
    for it in iter(dataloader):
        x, y = it
        x = x.to(device)
        y = y.to(device)
        sumlength += len(x)
        res += (torch.argmax(model.forward(x), dim=1) == torch.argmax(y, dim=1)).sum()
    model.train()
    return res / sumlength

config = {
    'epochs': 1,
    'learning_rate': 0.001,
    'batch_size': 64,
    'num_hidden': 1,
    'hidden_dim': 4,
    'limit': 128,
    'grid_size': 2,
    'spline_order': 2
}

# Create output directory
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
output_dir = f"./experiment/experiment_{timestamp}"
os.makedirs(output_dir, exist_ok=True)

# Save configuration
with open(os.path.join(output_dir, 'config.json'), 'w') as f:
    json.dump(config, f, indent=4)

# Use PCA dataloaders now
train_loader = get_pca_dataloader(split="train", batch_size=config['batch_size'], limit=-1)
val_loader = get_pca_dataloader(split="val", batch_size=config['batch_size'], limit=-1)
test_loader = get_pca_dataloader(split="test", batch_size=config['batch_size'])
ntk_loader = get_pca_dataloader(split="ntk", batch_size=config['batch_size'], limit=config['limit'])

# Initialize model with PCA input size and out_size=3 since we only have 3 classes now
model = ClassicKAN(
    num_hidden_layers=config['num_hidden'],
    hidden_dim=config['hidden_dim'],
    inp_size=n_components,  # PCA dimension
    out_size=3,
    grid_size=config['grid_size'],
    spline_order=config['spline_order']
).to(device)
model.net.to(device)

# Test a forward pass
for d,t in train_loader:
    print(d.shape)
    model.net(d.view(config['batch_size'], -1))
    break
print("out")

inp = [f"x_{i}" for i in range(n_components)]
model.net.plot(beta=100, scale=1, in_vars=inp, out_vars=["y_1","y_2","y_3"])
plt.show()

# Log model parameters
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"Total parameters: {total_params}")
print(f"Trainable parameters: {trainable_params}")

# Custom callback for logging
class MetricsCallback(L.Callback):
    def __init__(self, output_dir):
        super().__init__()
        self.metrics = {'train_loss': [], 'val_loss': []}
        self.output_dir = output_dir
        
    def on_train_epoch_end(self, trainer, pl_module):
        self.metrics['train_loss'].append(trainer.callback_metrics.get('train_loss', 0))
        
    def on_validation_epoch_end(self, trainer, pl_module):
        self.metrics['val_loss'].append(trainer.callback_metrics.get('val_loss', 0))
        
    def on_fit_end(self, trainer, pl_module):
        # Plot training curves
        plt.figure(figsize=(10, 6))
        plt.plot(self.metrics['train_loss'], label='Train Loss')
        plt.plot(self.metrics['val_loss'], label='Validation Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.title('Training Progress')
        plt.legend()
        plt.savefig(os.path.join(self.output_dir, 'training_curves.png'))
        plt.close()

# Training
metrics_callback = MetricsCallback(output_dir)
trainer = L.Trainer(
    max_epochs=config['epochs'],
    deterministic=True,
    callbacks=[metrics_callback],
    val_check_interval=1,
    enable_progress_bar=True
)

trainer.fit(model, train_loader, val_loader)
test_results = trainer.test(model, test_loader)

# Compute NTK accuracy
optimizer = torch.optim.Adam(model.parameters(), lr=config['learning_rate'])
ntk_model = GaussianFit(model=model, device=device, noise_var=0.0)
ntk_model.fit(ntk_loader, optimizer, MSELoss_batch)
ntk_acc = check_ntk_acc(model, test_loader)
print(f"NTK Accuracy: {ntk_acc.item()}")

# Save example predictions
example_batch = next(iter(test_loader))
example_images, example_labels = example_batch
predictions_KAN = model(example_images.to(device)).argmax(dim=1)
predictions_NTKAN = ntk_model(example_images.to(device)).argmax(dim=1)

# Since we have PCA vectors, we can't directly visualize them as images easily.
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i in range(min(10, len(example_images))):
    row = i // 5
    col = i % 5
    axes[row, col].imshow(np.zeros((7,7)), cmap='gray')
    axes[row, col].set_title(f'KAN: {predictions_KAN[i].item()}\nNTKAN: {predictions_NTKAN[i].item()}\nTrue: {torch.argmax(example_labels[i]).item()}')
    axes[row, col].axis('off')
plt.tight_layout()
plt.savefig(os.path.join(output_dir, 'predictions.png'))
plt.close()

# Save results
results = {
    'test_results': test_results,
    'ntk_accuracy': ntk_acc.item(),
    'total_parameters': total_params,
    'trainable_parameters': trainable_params
}
with open(os.path.join(output_dir, 'results.json'), 'w') as f:
    json.dump(results, f, indent=4)

print("Done.")
```
